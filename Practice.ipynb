```python
from datetime import datetime
row=df
def is_open(row):
    current_time = datetime.now().strftime("%H:%M")
    open_time = row["opening_time"]
    closing_time = row["closing_time"]
    if current_time <= open_time and current_time<=closing_time:
        return ("open")
        
    else :
        return("closed")
```

```python
df["Current time status"]=df.apply(is_open,axis=1)
```
